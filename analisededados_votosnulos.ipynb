```python
import pandas as pd
low_memory=False
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

Já no último código inserido, ele exibe as linhas e colunas que foram mencionadas anteriormemnte.

```python
df2 = df[['ANO_ELEICAO','QT_VOTOS_NULOS']].sort_values(by=("ANO_ELEICAO"), ascending=True).reset_index(drop=True)
plot = df2.plot(kind="bar", fontsize=(16), figsize=[22,6.5], x=("ANO_ELEICAO"), edgecolor="black")

for index, value in df2.iterrows():
    plot.text((index-1), (value["QT_VOTOS_NULOS"]+1500), str(value["QT_VOTOS_NULOS"]), fontsize=10, rotation=15)

plot.set_xlabel("ANO_ELEICAO", fontsize=(16));
plot.set_title("Quantidade de votos nulos em São José dos Campos - 2012 a 2018", fontsize=(20));
plot.set_ylabel("Quantidade", fontsize=(16));

plt.show()

print('O gráfico acima mostra a quantidade total de votos NULOS na cidade de São José dos Campos, em todas as eleições dos anos 2012 a 2018.')
print('Os múltiplos resultados apresentados devem-se a quantidade de zonas eleitorais e também ao tipo de eleição no respectivo ano, e também caso tenha ocorrido segundo turno.')
```

```python
df2 = df[['ANO_ELEICAO','QT_VOTOS_BRANCOS']].sort_values(by=("ANO_ELEICAO"), ascending=True).reset_index(drop=True)
plot = df2.plot(kind="bar", fontsize=(16), figsize=[22,6.5], x=("ANO_ELEICAO"), color="green", edgecolor="black")

for index, value in df2.iterrows():
    plot.text((index-1), (value["QT_VOTOS_BRANCOS"]+1500), str(value["QT_VOTOS_BRANCOS"]), fontsize=10, rotation=15)

plot.set_xlabel("ANO_ELEICAO", fontsize=(16));
plot.set_title("Quantidade de votos brancos em São José dos Campos - 2012 a 2018", fontsize=(20));
plot.set_ylabel("Quantidade", fontsize=(16));

plt.show()

print('O gráfico acima mostra a quantidade total de votos BRANCOS na cidade de São José dos Campos, em todas as eleições dos anos 2012 a 2018.')
print('Os múltiplos resultados apresentados devem-se a quantidade de zonas eleitorais e também ao tipo de eleição no respectivo ano, e também caso tenha ocorrido segundo turno.')
```

```python
df2 = df[['ANO_ELEICAO','QT_ABSTENCOES']].sort_values(by=("ANO_ELEICAO"), ascending=True).reset_index(drop=True)
plot = df2.plot(kind="bar", fontsize=(16), figsize=[22,8.5], x=("ANO_ELEICAO"), color="red", edgecolor="black")

for index, value in df2.iterrows():
    plot.text((index-1), (value["QT_ABSTENCOES"]+1500), str(value["QT_ABSTENCOES"]), fontsize=10, rotation=15)

plot.set_xlabel("ANO_ELEICAO", fontsize=(16));
plot.set_title("Quantidade de abstenções em São José dos Campos - 2012 a 2018", fontsize=(20));
plot.set_ylabel("Quantidade", fontsize=(16));

plt.show()

print('O gráfico acima mostra a quantidade total de ABSTENÇÕES na cidade de São José dos Campos, em todas as eleições dos anos 2012 a 2018.')
print('Os múltiplos resultados apresentados devem-se a quantidade de zonas eleitorais e também ao tipo de eleição no respectivo ano, e também caso tenha ocorrido segundo turno.')
```

```python
df.groupby('ANO_ELEICAO').sum().sort_values("ANO_ELEICAO", ascending=True).plot(kind='bar', grid=True, figsize=[16,5], title="Eleições em São José dos Campos", ylabel="Quantidade de votos em milhões")
```

O gráfico acima mostra resumidade a quantidade geral dos votos obtidos em cada ano, no intervalo de 2012 a 2018.

```python
df.groupby("ANO_ELEICAO")[["QT_VOTOS_NULOS", "QT_VOTOS_BRANCOS"]].sum().sort_values("ANO_ELEICAO", ascending=True).plot(kind='bar', grid=True, ylabel="Quantidade de votos")
```

O gráfico acima mostra resumidamente a quantidade de votos nulos e brancos obtidos em cada ano.

```python
df.groupby("ANO_ELEICAO")[["QT_APTOS", "QT_ABSTENCOES", "QT_VOTOS_NOMINAIS"]].sum().sort_values("ANO_ELEICAO", ascending=True).plot(kind='bar', grid=True, ylabel="Quantidade de votos")
```

O gráfico acima mostra resumidamente a quantidade geral de votos aptos, nominais e as abstenções obtidas em cada ano.
